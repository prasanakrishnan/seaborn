Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
#number of rows and cloumns
print('The Number of Row:',dataset.shape[0])
print('The number of column:',dataset.shape[1])
```

Univarite Analysis

```python
import seaborn as sns
sns.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})
```

Bivarite Analysis:Distribution plot

```python
sns.jointplot(x='ssc_p',y='salary',data=dataset)
plt.show()
```

```python
sns.jointplot(x='ssc_p',y='salary',data=dataset,kind='hex')
plt.show()
```

```python
sns.jointplot(x='ssc_p',y='salary',data=dataset,kind='kde')
plt.show()
```

```python
# Pair Plot
```

```python

sns.pairplot(dataset,hue = 'gender',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

For Categorical plot

```python

sns.stripplot(x = "ssc_b", y = "ssc_p", data = dataset)
plt.show()
```

```python
sns.stripplot(x = "gender", y = "ssc_p", data = dataset,jitter=True)
plt.show()
```

```python
sns.swarmplot(x = "gender", y = "ssc_p", data = dataset)
plt.show()
```

Distribution Observation

```python
sns.violinplot(x='gender',y='ssc_p',data=dataset)
plt.show()
```

```python
sns.violinplot(x='specialisation',y='salary',data=dataset)
plt.show()
```

```python
sns.violinplot(x='specialisation',y='salary',data=dataset,hue='gender')
plt.show()
```

```python
sns.boxplot(x='ssc_b',y='ssc_p',data=dataset,hue='gender')
plt.show()
```

Bar plot

```python
sns.barplot(x='ssc_b',y='ssc_p',data=dataset,hue='gender')
plt.show()
```

```python
sns.countplot(x='ssc_b',data=dataset,hue='gender')
plt.show()
```

```python
sns.pointplot(x='ssc_p',y='hsc_b',data=dataset,hue='gender')
plt.show()
```

```python
sns.boxplot(data=dataset,orient='h')
plt.show()
```
